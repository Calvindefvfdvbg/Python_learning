```python
# Get the number of products
size = int(input("Enter Product List Size: "))

# Dictionary to store product names and prices
products = {}

# Taking product names as input
for i in range(1, size + 1):
    product_name = input(f"Enter Product {i} -\n").strip().lower()
    price = input(f"Enter {product_name.capitalize()} Price:\n").strip()
    
    # Remove any non-numeric characters (like currency symbols)
    price = ''.join(filter(str.isdigit, price))
    
    # Store in dictionary after converting price to integer
    products[product_name] = int(price)

# Display the products and their prices
print("\n--- Output ---\n")
for product, price in products.items():
    print(f"{product} = {price}")
```

```python
-- User input section
Enter Product List Size : 4
Enter Product 1 -
Samosa
Enter Product 2 -
Kachori
Enter Product 3 -
Fafda
Enter Product 4 -
Jalebi

Enter Samosa Price :
300/-
Enter Kachori Price :
100/-
Enter Fafda Price :
100/-
Enter Jalebi Price :
200/-

Do You Want to add GST [Y/N]
[note : if user select `N` Print bill without GST]
[Note User select yes `Y`]
Enter GST % = 18
-------Out put -------

1. Samosa = 300/-
2. Kachori = 100/-
3. Fafda = 100/-
4. Jalebi = 200/-
-------------------
Total = 700
GST = 18%
-------------------
Final Total = 826/-
-------------------
```

```python
# Function to clean price input
def clean_price(price):
    return int(''.join(filter(str.isdigit, price)))  # Extracts only numbers

# Get product list size
size = int(input("Enter Product List Size: "))

# Dictionary to store product details
products = {}

# Taking product names and prices as input
for i in range(1, size + 1):
    product_name = input(f"Enter Product {i} -\n").strip().capitalize()
    price = input(f"Enter {product_name} Price:\n").strip()
    products[product_name] = clean_price(price)  # Store price as integer

# Ask if GST should be applied
gst_choice = input("Do You Want to add GST [Y/N]: ").strip().upper()

if gst_choice == 'Y':
    gst_percent = float(input("Enter GST % = ").strip())
else:
    gst_percent = 0

# Calculating total price
total = sum(products.values())
gst_amount = (total * gst_percent) / 100  # GST amount
final_total = round(total + gst_amount)  # Final bill

# Printing the bill
print("\n------- Output -------\n")

# Display the product list
for i, (product, price) in enumerate(products.items(), start=1):
    print(f"{i}. {product} = {price}/-")

print("-------------------")
print(f"Total = {total}")
if gst_choice == 'Y':
    print(f"GST = {gst_percent}%")
    print("-------------------")
    print(f"Final Total = {final_total}/-")
print("-------------------")
```
